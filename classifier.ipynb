```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, classification_report
db = pd.read_csv('diffData.csv')
db
```

```python
X = db[['time', 'distance']]
Y = db['Diffusion']
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

lr = LogisticRegression()
lr.fit(x_train, y_train)

y_pred = lr.predict(x_test)

conf_matrix = confusion_matrix(y_test, y_pred)
class_report = classification_report(y_test, y_pred)

print("Confusion Matrix:\n", conf_matrix)
print("\nClassification Report:\n", class_report)
```

```python
newX=(X-X.mean(axis = 0))/(X.std(axis = 0))
newX
```

```python
newx_train, newx_test, newy_train, newy_test = train_test_split(newX, Y, test_size=0.2, random_state=42)
lr.fit(newx_train, newy_train)
newy_pred = lr.predict(newx_test)

conf_matrix = confusion_matrix(newy_test, newy_pred)
class_report = classification_report(newy_test, newy_pred)

print("Confusion Matrix:\n", conf_matrix)
print("\nClassification Report:\n", class_report)
```
